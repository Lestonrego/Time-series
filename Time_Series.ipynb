```python

import matplotlib.pyplot as plt



data["Year"] = pd.to_datetime(data["Year"], format="%Y")


X=data[["Year"]]
y=data.set_index("Year")["Consumer price index (2010 = 100)"]
X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.96,shuffle=False)
model=ARIMA(y_train, order=(1,2,1))
result=model.fit()
start=len(y_train)
end=len(y_train)+len(y_test)-1
predictions=result.predict(start=start,end=end, typ="levels")
predictions.index = X_test["Year"].values


y_train.index = X_train["Year"].values
y_test.index = X_test["Year"].values
#inflation
test_inflation=predictions.pct_change()*100
plt.figure(figsize=(10,5))
y_train.plot(label="Train")
y_test.plot(label="Test", color="orange")
predictions.plot(label="Predictions", color="red",marker="*")
plt.axvline(y_test.index[0], color="gray", linestyle="--",label="TrainTest Split")
plt.legend()
plt.title("ARIMA Forecast")
plt.xlabel("Year")
plt.ylabel("CPI (2010=100)")
plt.show()
```

```python
results_df = pd.DataFrame({"Test": y_test,"Predictions": predictions,"Pre inflation":test_inflation})
print(results_df)
```

```python
from statsmodels.tools.eval_measures import rmse
fitted_vals=result.fittedvalues
fitted_vals.index=y_train.index
residuals=y_train-fitted_vals
# print(residuals)
# comparison=pd.DataFrame({"Actual":y_train,"Fitted":fitted_vals,"Residual":residuals})
# print(comparison)
trainr=rmse(y_train,fitted_vals)
print(trainr)
testr=rmse(y_test,predictions)

print(testr)
print("-----------")
diff = (trainr - testr)
print(diff)
```

```python
from sklearn.model_selection import train_test_split
from statsmodels.tools.eval_measures import rmse
import numpy as np

data["Year"] = pd.to_datetime(data["Year"],format="%Y")
X=data[["Year"]]
y=data.set_index("Year")["Consumer price index (2010 = 100)"]
train_sizes = [[0.60,0.1], [0.70,0.1],[ 0.80,0.1], [0.90,0.1]]
testa=[]
traina=[]
for sizes,size1 in train_sizes:
    X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=sizes,test_size=size1,shuffle=False)
    model=ARIMA(y_train, order=(1,2,2))
    result=model.fit()
    fitted_vals=result.fittedvalues
    fitted_vals.index=y_train.index
    trainr=rmse(y_train,fitted_vals)
    print(trainr)
    traina.append(trainr)  
    start=len(y_train)
    end=len(y_train)+len(y_test)-1
    predictions=result.predict(start=start,end=end)
    predictions.index=X_test["Year"].values
    y_train.index=X_train["Year"].values
    y_test.index=X_test["Year"].values
    #inflation
    test_inflation=predictions.pct_change()*100
    testr=rmse(y_test,predictions)
    testa.append(testr)
    print(testr)   
    print()
    # plt.figure(figsize=(10,5))
    # y_train.plot(label="Train")
    # y_test.plot(label="Test", color="orange")
    # predictions.plot(label="Predictions", color="red",marker="*")
    # plt.axvline(y_test.index[0], color="gray", linestyle="--",label="TrainTest Split")
    # plt.legend()
    # plt.title("ARIMA Forecast")
    # plt.xlabel("Year")
    # plt.ylabel("CPI (2010=100)")
    # plt.show()
print("mean of test",np.mean(testa))
print("mean of  train",np.mean(traina))
print()
print("std of test",np.std(testa))
print("std of train",np.std(traina))
```
