```python
# Cell 1 - Imports & Data
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA

warnings.filterwarnings("ignore")

# Load dataset
cpi = pd.read_csv("consumer price.csv")
country = "India"

# Filter for India and sort by Year
data = cpi[cpi["Entity"] == country].sort_values("Year")

# Use Year as index
y = data.set_index("Year")['Consumer price index (2010 = 100)']
print(y.head())
```

```python
# Cell 2 - Stationarity Test (choose d)
result = adfuller(y)
print("ADF Statistic:", result[0])
print("p-value:", result[1])

if result[1] < 0.05:
    print(" Series is stationary, try d=0")
else:
    print(" Series is not stationary, try differencing (d=1 or 2)")
```

```python
# First difference
y_diff1 = y.diff().diff().dropna()

# ADF test on differenced series
result = adfuller(y_diff1)
print("ADF Statistic (d=2):", result[0])
print("p-value (d=2):", result[1])

if result[1] < 0.05:
    print(" Series is stationary with d=1")
else:
    print(" Still not stationary, try d=2")
```

```python
import numpy as np
best_aic = np.inf
best_order = None
best_model = None

for p in range(0,10):
    for q in range(0, 10):
        try:
            model = ARIMA(y, order=(p, 2, q))  # <-- fixed d=2
            fit = model.fit()
            if fit.aic < best_aic:
                best_aic = fit.aic
                best_order = (p, 2, q)
                best_model = fit
        except:
            continue

print("Best ARIMA order:", best_order)
print("Best AIC:", best_aic)
```
