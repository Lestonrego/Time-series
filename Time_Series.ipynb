```python
import numpy as np
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tools.eval_measures import rmse
import warnings
warnings.filterwarnings("ignore")

# --- prepare data ---
data["Year"] = pd.to_datetime(data["Year"], format="%Y")
y = data.set_index("Year")["Consumer price index (2010 = 100)"]

# proportions
n = len(y)
train_fracs = [0.60, 0.70, 0.80, 0.90]
test_frac = 0.10

test_size = int(n * test_frac)
train_sizes = [int(n * frac) for frac in train_fracs]

print(f"Dataset length: {n}, test size: {test_size}, train sizes: {train_sizes}")

# Candidate ARIMA orders
candidates = [(1,1,0), (1,1,1), (1,1,2), (0,1,1)]

summary = []

for order in candidates:
    train_rmses = []
    test_rmses = []
    aics = []
    bics = []
    failed = False

    for size in train_sizes:
        if size + test_size > n:
            print(f"Not enough data for train {size} + test {test_size}")
            continue

        y_train = y.iloc[-(size + test_size):-test_size]
        y_test = y.iloc[-test_size:]

        try:
            model = ARIMA(y_train, order=order)
            res = model.fit()
        except Exception as e:
            print(f"Model {order} failed on train size {size}: {e}")
            failed = True
            break

        # AIC/BIC
        aics.append(res.aic)
        bics.append(res.bic)

        # Train RMSE
        try:
            fitted = res.fittedvalues
            fitted.index = y_train.index
            train_rmse = rmse(y_train, fitted)
        except Exception:
            train_preds = res.predict(start=0, end=len(y_train)-1)
            train_preds.index = y_train.index
            train_rmse = rmse(y_train, train_preds)

        # Test RMSE
        start = len(y_train)
        end = start + len(y_test) - 1
        test_preds = res.predict(start=start, end=end)
        test_preds.index = y_test.index
        test_rmse = rmse(y_test, test_preds)

        train_rmses.append(train_rmse)
        test_rmses.append(test_rmse)

    if failed or len(train_rmses) == 0:
        continue

    summary.append({
        "order": order,
        "train_rmse_mean": np.mean(train_rmses),
        "train_rmse_std": np.std(train_rmses, ddof=1),
        "test_rmse_mean": np.mean(test_rmses),
        "test_rmse_std": np.std(test_rmses, ddof=1),
        "aic_mean": np.mean(aics) if aics else np.nan,
        "bic_mean": np.mean(bics) if bics else np.nan
    })

# Print summary
print("\nModel evaluation summary:\n")
print("{:<10s} {:>12s} {:>12s} {:>12s} {:>12s} {:>12s} {:>12s}".format(
    "order", "tr_rmse_mean", "tr_rmse_std", "te_rmse_mean", "te_rmse_std", "aic_mean", "bic_mean"
))
for row in summary:
    print("{:<10s} {:12.4f} {:12.4f} {:12.4f} {:12.4f} {:12.2f} {:12.2f}".format(
        str(row["order"]),
        row["train_rmse_mean"],
        row["train_rmse_std"],
        row["test_rmse_mean"],
        row["test_rmse_std"],
        row["aic_mean"],
        row["bic_mean"]
    ))
```
